```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
from scipy.optimize import minimize

from fano import spike_stats
```

## Generate latent states

```python
n = 100000
alpha = 0.5 # higher alpha -> higher d' for Y on X
beta = 1.0 # higher beta -> higher d' for Y on X
ssq = 2.0 # higher ssq -> higher d' for Y on X
# mu = 1.0 # this will be set so that 0 < lambda < 1

# latent states
E = np.random.normal(0, np.sqrt(ssq), n)
X = np.zeros(n)
for i in xrange(1, n):
    X[i] = alpha*X[i-1] + E[i]
```

```python
plt.plot(X)
plt.title('latent variable')
plt.xlabel('time (msec)')
plt.ylabel('X(t)')
plt.show()

plt.hist(X, bins=100)
plt.xlabel('X')
plt.show()

plt.plot(X[:-1], X[1:], '.', alpha=0.5)
plt.title('X(t) | X(t-1)')
plt.xlabel('X(t-1)')
plt.ylabel('X(t)')
plt.show()
```

## Generate spike rates as draws from time-varying Bernoulli process

The following model is called an LNP model: Linear weighting of X by beta, Nonlinear rates given by the exponential, and Poisson spike generation approximated by a Bernoulli draw in each time bin).

```python
# spike rates, scaled so not more than 1 spike per bin
R0 = np.exp(beta*X)
mu = -np.log(R0.max())
R = np.exp(mu + beta*X)

Y = 1.0*(R > np.random.random(n))
# Y = np.random.poisson(R)

assert Y.max() < 2, Y.max()
print 'max spikes per bin = {0}\ntotal # spikes = {1}'.format(Y.max(), Y.sum())
mean, var = spike_stats(Y, 10)
print 'fano = {0}'.format(var/mean)
```

```python
plt.plot(R, '-', alpha=0.5)
plt.title('instantaneous spike rate, or probability of spiking in each bin')
plt.xlabel('time (msec)')
plt.ylabel('p')
plt.show()

plt.hist(np.log(R), bins=100)
plt.title('log spike rate distribution')
plt.xlabel('log(p)')
plt.ylabel('frequency')
plt.show()

plt.plot(R[:-1], R[1:], '.', alpha=0.5)
plt.title('R(t) | R(t-1)')
plt.xlabel('R(t-1)')
plt.ylabel('R(t)')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
plt.plot(X[Y==0], R[Y==0], '.b', alpha=0.5)
plt.plot(X[Y==1], R[Y==1], '.r', alpha=0.5)
plt.title('spike rate as a function of latent state')
plt.yscale('log')
plt.xlabel('X')
plt.ylabel('p')
plt.show()

plt.hist(X[Y==0], color='b', bins=20, alpha=0.5, normed=True)
plt.hist(X[Y==1], color='r', bins=20, alpha=0.5, normed=True)
plt.title('latent state conditional on spikes')
plt.xlabel('X')
plt.show()
```

## Estimate beta using MLE

```python
sX = X[Y==1].sum()
sY = Y.sum()
LL = lambda b: -(mu*sY + b*sX - np.sum(np.exp(mu + b*X)))
b0 = minimize(LL, 1.0, bounds=[(-100, 100)])['x']
bs = np.linspace(b0 - 1.0, b0 + 1.0)
plt.scatter(bs, [LL(b) for b in bs])
plt.scatter([b0], [LL(b0)], 50, c='r', label='MLE b = {0:.4f}'.format(b0[0]))
[c,d] = plt.ylim()
plt.plot([beta, beta], [c, d], '--', label='true b = {0}'.format(beta))
plt.title('MLE estimation of beta')
plt.xlabel('beta')
plt.ylabel('negative log-likelihood')
plt.legend(loc='upper left')
plt.show()
```

## Estimate latents and hyperparameters using MCMC

We want to sample from X and mu; assuming beta and ssq are known.

### Set hyperparameter priors

```python
# psi := exp(mu) ~ Gamma(v_p, w_p)
v_p = 1.0
w_p = 1.0
# alpha ~ N(v_a, w_a)
v_a = 1.0
w_a = 1.0
```

### Gibbs sample

```python
# M-H proposal covariance
ssq_mh = 1.0
V = np.diag(np.array([ssq_mh]*len(Y)))

# log-likelihood of Y given everything else
"""
P(y|lm) = (lm**y)*exp(-lm)
lg(P(y|lm)) = y*lg(lm) - lm
"""
LLY2 = lambda lglm, Y=Y: lglm[Y==1].sum() - np.exp(lglm).sum()
LLY = lambda x, alpha, beta, mu: LLY2(mu + beta*x)
# log-prior of X given everything else
"""
P(x) = P(x(t) | x(t-1)) * P(x(t-1) | x(t-2)) * ... * P(x(1))
P(x(t) | x(t-1)) = N(x(t) | alpha*x(t-1), ssq) [?]
n.b. implicitly setting prior on x[0] of 0
"""
LPX3 = lambda x, mu, cov=ssq: scipy.stats.normal(mu, np.sqrt(cov)).pmf(x)
LPX2 = lambda xs, mus, cov=ssq: np.log(np.array([LPX3(x, mu) for x, mu in zip(xs, mus)])).sum()
LPX = lambda x, alpha, beta, mu, Y=Y: LPX2(x[1:], alpha*x[:-1]) + np.log(LPX3(x[0], 0))

# starting guesses
X = np.zeros(len(Y))
mu = 0.2
alpha = 0.4
niters = 100

Xh = np.zeros([niters, len(Y)])
mh = np.zeros(niters)
ah = np.zeros(niters)

for i in xrange(niters):
    # P(X | Y, beta, mu, alpha)
    """
    -> M-H sampling of X all at once
    proposal (draw x*) = N(x, V)
        for some V...?
    likelihood := L(x) ≈ P(Y | x, beta, mu, alpha) * P(x | beta, mu, alpha)
    accept x* as new x with probability L(x*)/L(x)
    """
    Xstar = np.random.multivariate_normal(X, V)
    LL = lambda x, alpha=alpha, beta=beta, mu=mu: LLY(x, alpha, beta, mu) + LPX(x, alpha, beta, mu)
    if np.random.random() < np.exp(LL(Xstar) - LL(X)):
        X = Xstar
    
    # P(mu | Y, beta, X, alpha)
    """
    ≈ log(P(psi | Y, beta, X, alpha))
    ≈ P(Y | beta, X, alpha, psi) * P(psi | beta, X, alpha) [Bayes]
    = P(Y | beta, X, psi) * P(psi)
    = Poisson(psi*exp(X*beta)) * Gamma(v_p, w_p)
    = Gamma(v_p2, w_p2)
    where v_p2 := sum(Y) + v_p, and w_p2 := sum(exp(beta*X)) + w_p
    """
    v_p2 = Y.sum() + v_p
    w_p2 = np.exp(beta*X).sum() + w_p
    psi = np.random.gamma(v_p2, 1./w_p2)
    mu = np.log(psi)
    
    # P(alpha | Y, beta, X, mu)
    """
    -> regress X[1:] on X[:-1]
    = P(alpha | X)
    ≈ P(X | alpha) * P(alpha) [Bayes]
    = N(_, ssq) * N(v_a, w_a)
    := N(m, C)
    where 1/C := 1/w_a + xx/ssq, and m := C*(v_a/w_a + xy/ssq)
    where xx := sum(X[:-1]^2), xy := sum(X[1:]*X[:-1])
    """
    xx = (X[:-1]**2).sum()
    xy = (X[1:]*X[:-1]).sum()
    C = 1/(1/w_a + xx/ssq)
    m = C*(v_a/w_a + xy/ssq)
    alpha = np.random.normal(m, C)
    
    # save sample
    Xh[i,:] = X
    mh[i] = mu
    ah[i] = alpha
```
